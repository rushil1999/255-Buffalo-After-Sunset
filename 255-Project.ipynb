Removing all the rows having any of the relevant column entries to be null

```python
positional_columns = ['Longitude', 'Location', 'Latitude', 'Neighborhood', '2010 Census Tract', '2010 Census Block Group', '2010 Census Block',
                        'Census Tract', 'Census Block', 'Census Block Group',
                        'TRACTCE20', 'GEOID20_tract', 'GEOID20_blockgroup', 'GEOID20_block', 'Address']

indices_to_drop = pd.Index([])
for col in positional_columns:
  print(col)
  col_null_indices = data.loc[data[col].isnull()].index
  indices_to_drop = indices_to_drop.union(col_null_indices)

print(indices_to_drop)
data = data.drop(indices_to_drop)


```

```python
data.isnull().sum()
```

```python
data = data.drop(['updated_at', 'Incident ID', 'City', 'State'], axis=1)
data = data.reset_index()
```
