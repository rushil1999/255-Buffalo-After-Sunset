### Data Proprocessing

##### Handling Empty Columns

Column "Incident ID", "State", "updated_at" are completely empty columns to it should be removed from the dataset.

```python
df.drop(['Incident ID','State','updated_at'], inplace=True, axis=1)
df.head()
```

## Handling Add columns case by case starting from


1.   Case Number






As all the values in the *Case Number* column are disntinct it wont help us in infering the label class so we will remove that column.

```python
df.drop(['Case Number'], inplace=True, axis=1)
df.head()
```

2.  Incident Datetime

Column *Incident Datetime* consist of 2 feature, the date and the time.
This feature can be useful for understanding the crime relation so we will keep this in our dataset.

However its better that we represent this 2 feature independently.

```python
df['Incident Datetime'][0]
```

```python
def split_dt(x):
  date, time, am_pm = x.split()
  return date, time, am_pm
```

```python
df['Incident Datetime'] = df['Incident Datetime'].astype(str) 

df['Date'],df['time'],df['am_pm'] = df['Incident Datetime'].apply(split_dt,axis =1)
```
